## 조회할 때는 그냥 indexing / 값을 할당할때는 iat,at으로 하자

```python
####### append 방식으로 붙이기 #######

df_movie = pd.DataFrame() # 빈 dataframe 생성하기 -> 알아두자
df_movie_temp = pd.DataFrame()
# ignore_index = True : 칼럼명을 정렬
df_movie_temp = df_movie_temp.append({"movieCd" : "","movieNm" : "","movieNmEn" : "","openDt" : "","directorpeopleNm" : ""},ignore_index=True)

idx_moviecd = df_movie_temp.columns.get_loc("movieCd")
idx_movienm = df_movie_temp.columns.get_loc("movieNm")
idx_movienmen = df_movie_temp.columns.get_loc("movieNmEn")
idx_opendt = df_movie_temp.columns.get_loc("openDt")

for i in range(len(movieIdData["movieListResult"]["movieList"])):
    i_movie =  movieIdData["movieListResult"]["movieList"][i]
    df_movie_temp.iat[0,idx_moviecd] = i_movie["movieCd"]
    df_movie_temp.iat[0,idx_movienm] = i_movie["movieNm"]
    df_movie_temp.iat[0,idx_movienmen] = i_movie["movieNmEn"]
    df_movie_temp.iat[0,idx_opendt] = i_movie["openDt"]
    if len(i_movie["directors"]) != 0:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = i_movie["directors"][0]["peopleNm"]
    else:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = ""
    
    # 여기서 append해서 다시 df_movie에 추가하는 것은 바로 df_movie에서 iat indexing으로 쌓는 것이 막혀서 그렇다.
    df_movie = df_movie.append(df_movie_temp, ignore_index=True)    
    
    # 다시 빈 dataframe값으로 돌리는 것이 좋다. 왜냐하면 앞에서 했던 값들이 남아있을 수 있기 때문이다.
    df_movie_temp.iloc[0] = ""
    
df_movie
```

```python
####### concat 함수 방식으로 붙이기 #######

df_movie = pd.DataFrame() # 빈 dataframe 생성하기 -> 알아두자
df_movie_temp = pd.DataFrame()
# ignore_index = True : 칼럼명을 정렬
df_movie_temp = df_movie_temp.append({"movieCd" : "","movieNm" : "","movieNmEn" : "","openDt" : "","directorpeopleNm" : ""},ignore_index=True)

idx_moviecd = df_movie_temp.columns.get_loc("movieCd")
idx_movienm = df_movie_temp.columns.get_loc("movieNm")
idx_movienmen = df_movie_temp.columns.get_loc("movieNmEn")
idx_opendt = df_movie_temp.columns.get_loc("openDt")

for i in range(len(movieIdData["movieListResult"]["movieList"])):
    i_movie =  movieIdData["movieListResult"]["movieList"][i]
    df_movie_temp.iat[0,idx_moviecd] = i_movie["movieCd"]
    df_movie_temp.iat[0,idx_movienm] = i_movie["movieNm"]
    df_movie_temp.iat[0,idx_movienmen] = i_movie["movieNmEn"]
    df_movie_temp.iat[0,idx_opendt] = i_movie["openDt"]
    if len(i_movie["directors"]) != 0:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = i_movie["directors"][0]["peopleNm"]
    else:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = ""
    
    # concat 방식으로 하기
    df_movie = pd.concat([df_movie,df_movie_temp], ignore_index = True)
    #df_movie = df_movie.append(df_movie_temp, ignore_index=True)    
   
    # 다시 빈 dataframe값으로 돌리는 것이 좋다. 왜냐하면 앞에서 했던 값들이 남아있을 수 있기 때문이다.
    df_movie_temp.iloc[0] = ""
    
df_movie.head(10)
```

## lambda, apply를 이용해서 개봉날짜 연, 월, 일로 나누기
#### lambda 적용할 때는 항상 x 대상이 무엇인지를 생각하자

```python
df_movie["year"] = df_movie["openDt"].apply(lambda x: str(x)[:4])
df_movie["month"] = df_movie["openDt"].apply(lambda x: str(x)[4:6])
df_movie["day"] = df_movie["openDt"].apply(lambda x: str(x)[6:])
df_movie.head(10)
```

## data frame을 원하는 대로 컬럼 순서 변경 및 index 설정

```python
# 참고) 이미 set_index 한 이후에 다시 하게 되면 에러 발생
df_movie = df_movie.sort_values(by = "openDt") # by = ["A","B"] 이런식으로 여러개를 기준으로 sorting가능
df_movie = df_movie.set_index("movieCd")

# 원하는 순서대로 컬럼 변경
# indexing 안에 list형식으로 해서 원하는 순서로 정렬하면 된다.
df_movie = df_movie[["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNm"]]
df_movie.head(10)
```

## data frame으로 만들기

```python
df_movie["type"] = ""
df_movie["ActorNum"] = ""
df_movie["cast"] = ""
df_movie["cast_peopleNm"] = ""

# 칼럼이름에 대한 정수 위치
for i in range(0, len(df_movie)):
    movieCode = df_movie.index[i]
    movieDetailedURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=0566dcfb243b0d5faf5359937f2176a1&movieCd="+ movieCode
    movieIdDetailedPage = urllib.request.urlopen(movieDetailedURL)
    
    movieIdDetailed = json.loads(movieIdDetailedPage.read().decode('utf8'), encoding='utf-8')
    
    df_movie.iat[i, df_movie.columns.get_loc("type")] = movieIdDetailed["movieInfoResult"]["movieInfo"]["typeNm"]
    df_movie.iat[i, df_movie.columns.get_loc("ActorNum")] = len(movieIdDetailed["movieInfoResult"]["movieInfo"]["actors"])
    if len(movieIdDetailed["movieInfoResult"]["movieInfo"]["actors"]) != 0:
        df_movie.iat[i, df_movie.columns.get_loc("cast")] = movieIdDetailed["movieInfoResult"]["movieInfo"]["actors"][0]["cast"]
        df_movie.iat[i, df_movie.columns.get_loc("cast_peopleNm")] = movieIdDetailed["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"]
    else:
        df_movie.iat[i, df_movie.columns.get_loc("cast")] = ""
        df_movie.iat[i, df_movie.columns.get_loc("cast_peopleNm")] = ""
        
df_movie.head(10)
```

## XML을 이용해서 위에서 작업했던 것처럼 dataframe 만들기

```python
df_xml_movie = pd.DataFrame() # 빈 dataframe 생성하기 -> 알아두자
df_xml_movie_temp = pd.DataFrame()
# ignore_index = True : 칼럼명을 정렬
df_xml_movie_temp = df_xml_movie_temp.append({"movieCd" : "","movieNm" : "","movieNmEn" : "","openDt" : "","directorpeopleNm" : ""},ignore_index=True)

for i in range(0,len(soup.find_all("moviecd"))):
    df_xml_movie_temp.iat[0, df_xml_movie_temp.columns.get_loc("movieCd")] = soup.find_all("moviecd")[i].text
    df_xml_movie_temp.iat[0, df_xml_movie_temp.columns.get_loc("movieNm")] = soup.find_all("movienm")[i].text
    df_xml_movie_temp.iat[0, df_xml_movie_temp.columns.get_loc("movieNmEn")] = soup.find_all("movienmen")[i].text
    df_xml_movie_temp.iat[0, df_xml_movie_temp.columns.get_loc("openDt")] = soup.find_all("opendt")[i].text
    df_xml_movie_temp.iat[0, df_xml_movie_temp.columns.get_loc("directorpeopleNm")] = soup.find_all("directors")[i].text

    df_xml_movie = df_xml_movie.append(df_xml_movie_temp, ignore_index=True)
    df_xml_movie_temp.iloc[0] = ""
    
df_xml_movie
```

```python
df_xml_movie["year"] = df_xml_movie["openDt"].apply(lambda x: str(x)[:4])
df_xml_movie["month"] = df_xml_movie["openDt"].apply(lambda x: str(x)[4:6])
df_xml_movie["day"] = df_xml_movie["openDt"].apply(lambda x: str(x)[6:])
df_xml_movie
```

```python
# 참고) 이미 set_index 한 이후에 다시 하게 되면 에러 발생
df_xml_movie = df_xml_movie.sort_values(by = "openDt") # by = ["A","B"] 이런식으로 여러개를 기준으로 sorting가능
df_xml_movie = df_xml_movie.set_index("movieCd")

# 원하는 순서대로 컬럼 변경
df_xml_movie = df_xml_movie[["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNm"]]
df_xml_movie
```

```python
df_xml_movie["type"] = ""
df_xml_movie["ActorNum"] = ""
df_xml_movie["cast"] = ""
df_xml_movie["cast_peopleNm"] = ""

for code in df_xml_movie.index:
    movieDetailedURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?key=0566dcfb243b0d5faf5359937f2176a1&movieCd="+ code 
    movieIdDetailedPage = urllib.request.urlopen(movieDetailedURL)
    
    soup_detailed = BeautifulSoup(movieIdDetailedPage, "html.parser")
    
    df_xml_movie.at[code, "type"] = soup_detailed.find("typenm").text
    df_xml_movie.at[code, "ActorNum"] = len(soup_detailed.find_all("actor"))
    if len(soup_detailed.find_all("actor")) != 0:
        df_xml_movie.at[code, "cast"] = soup_detailed.actors.find("cast").text
        df_xml_movie.at[code, "cast_peopleNm"] = soup_detailed.actors.find("peoplenm").text
    else:
        df_xml_movie.at[code, "cast"] = ""
        df_xml_movie.at[code, "cast_peopleNm"] = ""
    
    
df_xml_movie
```
