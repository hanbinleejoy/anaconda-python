# 06. 3) Pivot

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../data/11_sales-funnel.xlsx")
```

```python
data.head()
```

```python
# 좀더 명확하게 하기 위해서 카테고리 변수화
# R에서 factor화 시키는 것
data["Status"] =  data["Status"].astype("category")

data["Status"].cat.set_categories(
    ["won","pending","presented","declined"],
    inplace=True)

data
```

### pivot table 표현하기

```python
# Name을 가로줄 기준(행)으로 하고. 데이터를 보자 -> 숫자인 칼럼만 들어온다.
pd.pivot_table(data, index=["Name"]).head()
```

```python
# 기준을 세가지로 할 때
# 각 그룹에 해당하는 데이터들의 평균치로 보여둔다.
pd.pivot_table(data, index=["Name","Rep","Manager"]).head()
```

```python
# Manager : 담당 매니저
# Rep : 영업사원
pd.pivot_table(data, index=["Manager","Rep"])
# 평균은 outlier에 상당히 민감하다.
# 이것만으로 (누가 더 뛰어난지)정확한 판단을 하기에는 무리
```

```python
# Account 정보는 필요가 없기에 제외
pd.pivot_table(data,
               index=["Manager","Rep"],
               values=["Price"])
```

```python
# 평균이 아닌 총합 표시
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price"],
              aggfunc=np.sum)
```

```python
# 여러개의 지표를 보이고 싶을 때 aggfunc에 list로 넘기면 된다.
pd.pivot_table(data, index=["Manager","Rep"],
              values=["Price"],
              aggfunc=[np.mean, len])
```

```python
# 세로줄에 대한 계층
pd.pivot_table(data, index=["Manager","Rep"],
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum])
```

```python
# 결측치를 0으로 초기화
pd.pivot_table(data, index=["Manager","Rep"],
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
pd.pivot_table(data, index=["Manager","Rep"],
               values=["Price","Quantity"],
               columns=["Product"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
# 품목에 대한 정보를 가로 인덱스로
pd.pivot_table(data, index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
pd.pivot_table(data, index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum, np.mean],
               fill_value=0)
```

```python
# row : manager / status
# column : Quantity [len] / Price [sum]
# column category : product
pd.pivot_table(data, index=["Manager","Status"],
               columns=["Product"],
               values=["Quantity","Price"],
               aggfunc={"Quantity":len, "Price":np.sum},
               fill_value=0)
```

```python
pd.pivot_table(data, index=["Manager","Status"],
               columns=["Product"],
               values=["Quantity","Price"],
               aggfunc={"Quantity":len, "Price":[np.sum, np.mean]},
               fill_value=0)
```

```python
# pivot & filter
table = pd.pivot_table(data, index=["Manager","Status"],
               columns=["Product"],
               values=["Quantity","Price"],
               aggfunc={"Quantity":len, "Price":[np.sum, np.mean]},
               fill_value=0)

table.query("Manager == ['Debra Henley'] & Status == ['won']")
```

### crosstab
pd.crosstab([],[])
#### 오로지 빈도에 대한 결과값을 다룬다.

```python
#crosstab
pd.crosstab(data["Name"], data["Price"])
# unique한 값을 뽑아서 작성해준다.
```

```python
# 가로와 세로에 여러개의 칼럼변수를 가지고 보여주고 싶을 때
pd.crosstab([data["Manager"], data["Rep"]],
            [data["Price"], data["Quantity"]])
```
