# 06. 2) join

```python
import pandas as pd
```

```python
baseball_player = pd.DataFrame(
    {
        "player_Name": ["권혁","이범호","필","박병호","양준혁"],
        "team_No": [1993, 2001, 2001, 2010, None]
    }
)

baseball_team = pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    }
)
```

```python
baseball_player
```

```python
baseball_team
```

## Inner Join
### left Join: 왼쪽테이블 전체에서 오른쪽 테이블에서 
값을 찾아 가져올 수 있는 것만 가져오고 못가져오는 건 NaN

```python
# Left Outter Join
pd.merge(baseball_player, baseball_team, how='left',
         left_on="team_No", right_on="team_id")
```

#### right join: left와 반대로 오른쪽을 기준으로 다 가져오고 없는 데이터는 NaN

```python
pd.merge(baseball_player, baseball_team, how='right',
         left_on="team_No", right_on="team_id")
```

## Outer Join
### left join
순수하게 left

```python
# left excluding join
# 01) inner left join
temp = pd.merge(baseball_player, baseball_team,
               how='left',
               left_on = 'team_No',
               right_on='team_id')

# 02) select NaN
temp[temp['team_No'].isnull()]
```

### right join

```python
# right excluding join
# 01) inner right join
temp = pd.merge(baseball_player, baseball_team,
               how='right',
               left_on = 'team_No',
               right_on='team_id')

# 02) select NaN
temp[temp['player_Name'].isnull()]
```

### 데이터 합치기

```python
# 그냥 left, right 두개 데이터 싹다 합치기
pd.merge(baseball_player, baseball_team,
               how='outer',
               left_on = 'team_No',
               right_on='team_id')
```

#### 선수만 있거나 팀 정보만 있는 데이터 (A,B 교집합만 빼고 나머지)

```python
temp = pd.merge(baseball_player, baseball_team,
               how='outer',
               left_on = 'team_No',
               right_on='team_id')

# left join에서 빠진 부분과 right join에서 빠진 부분 합치기
temp[temp["player_Name"].isnull() | temp["team_No"].isnull()]
```

## df에 index가 존재할 때의 merge

```python
baseball_player = pd.DataFrame(
    {
        "player_Name": ["권혁","이범호","필","박병호","양준혁"],
        "team_No": [1993, 2001, 2001, 2010, None]
    },
    index = ["p_01","p_02","p_03", "p_04","p_05"]
)

baseball_player.index.name = "player_id"
baseball_player
```

```python
baseball_team = pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
    },
    index = [1993,1999,2001,2010]
)
baseball_team
```

```python
# Left Outer Join
temp = pd.merge(baseball_player, baseball_team,
               how='left',
               left_on = 'team_No', # column명
               right_index=True) # right을 index기준으로 할 것인가

temp[temp["team_No"].isnull()]
```

```python
# Right Outer Join
temp = pd.merge(baseball_player, baseball_team,
               how='right',
               left_on = 'team_No', # column명
               right_index=True) # right을 index기준으로 할 것인가

temp[temp["player_Name"].isnull()]
```

```python
# Outer
temp = pd.merge(baseball_player, baseball_team,
               how='outer',
               left_on = 'team_No', # column명
               right_index=True) # right을 index기준으로 할 것인가

temp[temp["player_Name"].isnull() | temp["team_No"].isnull()]
```

```python
# Inner (겹치는 데이터만 가져옴)
pd.merge(baseball_player, baseball_team,
               how='inner',
               left_on = 'team_No', # column명
               right_index=True) # right을 index기준으로 할 것인가
```
