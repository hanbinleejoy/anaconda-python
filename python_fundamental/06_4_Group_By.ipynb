# 06. 4) Group_By

```python
import pandas as pd
import numpy as np
import dateutil
```

```python
# 과거
# data = pd.read_csv("data/10_phone_data.csv", parse_date=True)

data = pd.read_csv("data/10_phone_data.csv")
data['date'] = data['date'].apply(dateutil.parser.parse, dayfirst=True)
```

```python
data.head(10)
```

```python
print(data["item"].count()) #len(data) 와 같다.
print(data["item"].value_counts())
# item칼럼 별로 몇 개가 존재하는지
```

```python
# item이 call인 경우의 총 du 시간
data['duration'][data['item'] == 'call'].sum()
```

```python
# 가장 긴 데이터 사용시간
print(data["duration"].max())
```

```python
# month로 묶어보자 -> 바로 보이지가 않음
data.groupby(['month'])
```

```python
# 묶은 것의 키를 보자
data.groupby(['month']).groups.keys()
```

```python
data.groupby(['month']).groups.values()
```

```python
data.groupby(['month']).groups['2014-11']
```

```python
# 각 데이터를 보는 방법
for name, group in data.groupby('month'):
    print(name)
    print(group)
```

```python
# 묶은 후에 다른 함수 적용하기
# head()하면 각 그룹별로 head()
data.groupby('month').head()
```

```python
# 데이터를 그룹화시키고 각 그룹별 평균 구하기 (숫자데이터가 duration밖에 없음)
data.groupby('month').mean()
```

```python
# 직접 묶은 후에 연산을 적용할 칼럼을 지정하고 수행
data.groupby('month')['duration'].sum()
```

```python
data.groupby('month')
```

```python
data.groupby('month')['date'].count()
# data['month'].value_counts() 
# 같은 것임
```

```python
data[data['item'] == 'call'].groupby('network')['duration'].sum()
```

```python
# 여러 조건으로 그룹바이가능
data.groupby(['month','item'])['date'].count()
```

```python
data.groupby(['month','item']).agg({'duration':sum, 'network_type':'count','date':'first'})

# 위와 똑같다.
# pd.pivot_table(data, index=['month','item'],
#              aggfunc={'duration':sum,'network_type':'count','date':'first'})
```
