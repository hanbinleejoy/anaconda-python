## KOBIS 데이터 실습

```python
import pandas as pd
import urllib.request
import json
```

```python
# JSON Format

movieURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=0566dcfb243b0d5faf5359937f2176a1&openStartDt=2020&itemPerPage=40"
movieIdPage = urllib.request.urlopen(movieURL)

# 참고)
#print(movieIdPage)
#print(movieIdPage.read())
#print(movieIdPage.read().decode("utf8"))

movieIdData = json.loads(movieIdPage.read().decode('utf8'), encoding='utf-8')
movieIdData
```

```python
len(movieIdData["movieListResult"]["movieList"]) # 40개 영화목록 요청 완료
```

```python
movieIdData["movieListResult"]["movieList"][0]["directors"][0]["peopleNm"] # 영화 감독 이름 출력
```

```python
for i in range(len(movieIdData["movieListResult"]["movieList"])):
    print("Code :", movieIdData["movieListResult"]["movieList"][i]["movieCd"])
```

```python
for i in range(len(movieIdData["movieListResult"]["movieList"])):
    print(str(i+1)+"번째 영화")
    print("Code :", movieIdData["movieListResult"]["movieList"][i]["movieCd"])
    if len(movieIdData["movieListResult"]["movieList"][i]["directors"]) != 0:
        print("Director :", movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"])
    else:
        print("Director : 정보가 없습니다.")
```

### 조회할 때는 그냥 indexing / 값을 할당할때는 iat,at으로 하자

```python
df_movie = pd.DataFrame() # 빈 dataframe 생성하기 -> 알아두자
df_movie_temp = pd.DataFrame()
# ignore_index = True : 칼럼명을 정렬
df_movie_temp = df_movie_temp.append({"movieCd" : "","movieNm" : "","movieNmEn" : "","openDt" : "","directorpeopleNm" : ""},ignore_index=True)


idx_moviecd = df_movie_temp.columns.get_loc("movieCd")

for i in range(len(movieIdData["movieListResult"]["movieList"])):
    i_movie =  movieIdData["movieListResult"]["movieList"][i]
    df_movie_temp.iat[0,idx_moviecd] = i_movie["movieCd"]
    df_movie_temp.iat[0,df_movie_temp.columns.get_loc("movieNm")] = i_movie["movieNm"]
    df_movie_temp.iat[0,df_movie_temp.columns.get_loc("movieNmEn")] = i_movie["movieNmEn"]
    df_movie_temp.iat[0,df_movie_temp.columns.get_loc("openDt")] = i_movie["openDt"]
    if len(i_movie["directors"]) != 0:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = i_movie["directors"][0]["peopleNm"]
    else:
        df_movie_temp.iat[0,df_movie_temp.columns.get_loc("directorpeopleNm")] = ""
    # 여기서 append해서 다시 df_movie에 추가하는 것은 바로 df_movie에서 iat indexing으로 쌓는 것이 막혀서 그렇다.
    df_movie = df_movie.append(df_movie_temp, ignore_index=True)    
    # 다시 빈 dataframe값으로 돌리는 것이 좋다. 왜냐하면 앞에서 했던 값들이 남아있을 수 있기 때문이다.
    df_movie_temp.iloc[0] = ""
    
df_movie
```
